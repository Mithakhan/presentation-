## Processing stopwords

```python
# read stopwords from file
with open("stopwords-ur.txt", 'r', encoding='utf8') as f:
    data = f.read().replace("\n", " ")

# removing accents
rm_acc = remove_accents(data)

# normalize whitespace
rm_whitespc = normalize_whitespace(rm_acc)

# splitting into each document as a string
stopwords = rm_whitespc.split(',') 
```

```python
# create blank document
word_blank_page = blank('ur')

# place dataset into document
doc = word_blank_page(str(stopwords))
```

```python
# convert doc to str
text = doc.text

# clean text
text = text.replace('[', '')
text = text.replace(']', '')
text = text.replace("'", '')

# convert to list of stopwords
stopwords = text.split(' ')
```

## Removing stopwords from dataset (words)

```python
final_words = [str(word)   for word in words   if word not in stopwords]

print("Words: {}, stopwords: {}".format(len(words), len(words) - len(final_words)))

print(final_words)
```
